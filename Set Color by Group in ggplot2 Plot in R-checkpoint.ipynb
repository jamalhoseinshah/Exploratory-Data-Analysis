```r
set.seed(68117534)              # Set random seed
data <- data.frame(x = 1:15,    # Create example data frame
                   y = c(runif(15, 0, 2),
                         runif(15, 2, 6),
                         runif(15, 3, 4),
                         runif(15, 1, 3)),
                   group = rep(LETTERS[1:4], each = 15))
```

```r
head(data)                      # Print head of example data frame
```

```r
install.packages("ggplot2")     # Install & load ggplot2 package
```

```r
library("ggplot2")
```

```r
ggplot(data,                    # Draw ggplot2 plot without colors
       aes(x = x,
           y = y,
           group = group)) +
  geom_line() +
  geom_point()
```

```r
# Modify Colors of Single Geom by Group
```

```r
ggplot(data,                    # Change colors of lines by group
       aes(x = x,
           y = y,
           group = group)) +
  geom_line(aes(col = group)) +
  geom_point()
```

```r
# Modify Colors of All Geoms by Group
```

```r
ggplot(data,                    # Change colors of lines & points by group
       aes(x = x,
           y = y,
           col = group)) +
  geom_line() +
  geom_point()
```

```r
# Using Manual Color Codes by Group
```

```r
ggplot(data,                    # Specify colors manually
       aes(x = x,
           y = y,
           col = group)) +
  geom_line() +
  geom_point() +
  scale_color_manual(values = c("#ca7dcc",
                                "#1b98e0",
                                "#353436",
                                "#02e302"))
```

```r
# Modify Fill Colors of Boxplots by Group
```

```r
ggplot(data,                    # Change colors of boxplots
       aes(y = y,
           fill = group)) +
  geom_boxplot()
```
